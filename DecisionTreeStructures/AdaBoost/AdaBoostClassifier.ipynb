```python
x=df.drop('Outcome', axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=15)
```

```python
columns_to_fill= ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
```

```python
medians={}
for col in columns_to_fill :
  median_value = X_train[X_train[col] != 0][col].median()
  medians[col] = median_value
  X_train[col] = X_train[col].replace(0, median_value)

for col in columns_to_fill:
  X_test[col] = X_test[col].replace(0, medians[col])
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
#bu ağaçla alakalı bir algoritma bu yüzden scaler a gerek yok
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
ada = AdaBoostClassifier() 
#zayıf tahmincileri (genellikle derinliği sadece 1 olan karar ağaçları) bir araya getirerek çok güçlü bir model oluşturmaktır.
#estimator: Hangi zayıf modeli kullanacağındır , n_estimators: Kaç tane zayıf modelin art arda eğitileceğidir
#learning_rate: Her bir modelin genel karara ne kadar etki edeceğini belirler, 1000 gibi büyük sayılar verirsek overfittinge sebep olabilir
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

adaboost_param = {
        "n_estimators" : [50, 70, 100, 120, 150, 200],
        "learning_rate" : [0.001, 0.01, 0.1, 1, 10] #1000 gibi büyük sayılar verirsek overfittinge sebep olabilir
}
```

```python
grid=GridSearchCV(estimator=AdaBoostClassifier() , param_grid=adaboost_param, cv=3 , verbose=1 , n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
ada = AdaBoostClassifier(learning_rate=1, n_estimators=150)
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
