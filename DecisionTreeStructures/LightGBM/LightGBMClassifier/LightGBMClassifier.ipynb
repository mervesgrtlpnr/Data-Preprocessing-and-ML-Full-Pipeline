feature engineering

```python
df.isnull().sum()
```

```python
#embar_town, deck ve alive ' sütunlarını silelim
```

```python
df=df.drop(['embark_town', 'deck', 'alive'], axis=1)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['age']=df['age'].fillna(df['age'].median())
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0]) #embarked da en çok tekrar eden değerle doldurduk
```

```python
df.info()
```

```python
df.sex.value_counts()
```

```python
df['class'].value_counts()
```

```python
df['embarked'].value_counts()
```

```python
df['who'].value_counts()
```

```python
df['adult_male'] = df['adult_male'].astype(int)
df['alone'] = df['alone'].astype(int)
```

```python
X = df.drop('survived', axis=1)
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 15)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['sex', 'class', 'who','embarked']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'  
)

X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
encoded_cols = preprocessor.get_feature_names_out()
```

```python
X_train = pd.DataFrame(X_train, columns=encoded_cols)
X_test = pd.DataFrame(X_test, columns=encoded_cols)
```

```python
import lightgbm as lgb
from sklearn.metrics import classification_report, confusion_matrix
clf = lgb.LGBMClassifier(verbosity=-1) #sessiz modda çalışır ,uyarıları çıktıları göstermez
clf.fit(X_train, y_train)
#LGBMClassifier kayıptaki azalmanın en çok olduğu yaprağı seçer ve oradan derinleşir bu sayede çok daha az hatayla daha hızlı öğrenir
```

```python
y_pred = clf.predict(X_test)
```

```python
print(confusion_matrix( y_pred , y_test))
print(classification_report( y_pred, y_test))
```

```python
clf.feature_importances_
X_train.columns
```

```python
importances = clf.feature_importances_

feature_names = X_train.columns
feat_importance = pd.DataFrame({
    'Feature': feature_names,  #birinci sütun özellik adları
    'Importance': importances  #ikinci sütun ise önem skorları
}).sort_values(by='Importance', ascending=False)

print(feat_importance)

#feature_importances_:Model eğitimi bittikten sonra,
#her bir sütunun sonucu tahmin etmede ne kadar etkili olduğunu 0 ile 1 arasında bir skorla bu değişkene kaydeder
```

LightGBM'de hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV

lgb_model = lgb.LGBMClassifier(verbosity=-1)

param_grid = {
    'n_estimators': [100, 300, 500, 1000],
    'max_depth': [3, 5, 7, -1],
    'learning_rate': [0.01, 0.05, 0.1, 0.3],
    'num_leaves': [15, 31, 63, 127],
    'min_child_samples': [5, 10, 20],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

# Randomized search
random_search = RandomizedSearchCV(
    estimator=lgb_model,
    param_distributions=param_grid,
    scoring='accuracy',
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV

lgb_model = lgb.LGBMClassifier(verbosity=-1)

param_grid = {
    'n_estimators': [100, 300, 500, 1000], #kaç tane ağaç oluşturulacağını belirler
    'max_depth': [3, 5, 7, -1], #ağaçların maksimum derinliğini belirler, -1 değeri sınırsız derinlik anlamına gelir
    'learning_rate': [0.01, 0.05, 0.1, 0.3], #tahmine olan katkıyı belirler, daha düşük değerler daha yavaş öğrenmeye neden olur 
    'num_leaves': [15, 31, 63, 127],#2 üssü -1 yaprak sayısını belirler, daha fazla yaprak daha karmaşık bir model anlamına gelir
    'min_child_samples': [5, 10, 20], #bir yaprağın bölünebilmesi için gereken min örnek sayısını belirler yüksek değerler daha basit modeller oluşturur
    'subsample': [0.6, 0.8, 1.0], #toplam verinin yüzde kaçını her ağaç için kullanacağını belirler
    'colsample_bytree': [0.6, 0.8, 1.0] #her ağacı oluştururken kullanılacak özelliklerin oranını belirler
}

# Randomized search
random_search = RandomizedSearchCV(
    estimator=lgb_model,
    param_distributions=param_grid,
    scoring='accuracy',
    cv=5,
    verbose=1, #en azını göstersin
    random_state=42,
    n_jobs=-1 #hızlı olsun diye -1
)

random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```

```python
y_pred= random_search.predict(X_test)
```

```python
print(confusion_matrix( y_pred , y_test))
print(classification_report( y_pred, y_test))
```
