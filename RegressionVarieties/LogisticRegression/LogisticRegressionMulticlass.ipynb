```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/7-cyber_attack_data.csv')
```

```python
df.columns
```

```python
#src_packet_rate (Kaynak tarafı paket gönderim oranı): Bir bağlantı boyunca kaynak tarafından saniyede gönderilen paketlerin oranını ifade eder.(ne kadar paket geldiği)

#dst_packet_rate (Hedef tarafı paket alım oranı):Hedef sistemin saniyede aldığı paket sayısını gösterir.(gelen isteğin ne kadarını aldığımızı)

#avg_payload_size (Ortalama yük boyutu):iletilen paketlerin veri içeriğinin ortalama büyüklüğünü belirtir.(isteğin büyüklüğü)

#connection_duration (Bağlantı süresi):İki uç arasındaki bağlantının saniye cinsinden toplam süresidir.(isteğin kac sn ne kadar sürdüğü)

#tcp_flag_count (TCP bayrak sayısı):TCP protokolündeki belirli bayrakların (örneğin SYN, ACK vb.) kaç kez görüldüğünü ifade eder.

#avg_interarrival_time (Ortalama paket arası süre):Paketlerin birbirini takip etme sürelerinin ortalamasını verir.(istekler arasındaki süre)

#failed_login_attempts (Başarısız giriş denemeleri sayısı):Kullanıcıların başarısız oturum açma girişimlerinin sayısını gösterir.

#unusual_port_activity_score (Olağandışı port etkinliği skoru):Sistemde alışılmışın dışında port kullanımı olup olmadığını yansıtır.

#session_entropy (Oturum entropisi):Oturum davranışının rastgelelik veya düzensizlik derecesini ölçer ve anomali tespitinde kullanılır.

#avg_response_delay (Ortalama yanıt gecikmesi):Sunucunun isteklere yanıt verme süresinin milisaniye cinsinden ortalamasını ifade eder.

#attack_type (Saldırı türü):Trafiğin normal mi yoksa saldırı kaynaklı mı olduğunu belirtir.

#0 → Normal  1 → DDoS saldırısı  2 → Port taraması(Port Scan)
```

```python
df.describe()
```

```python
X=df.drop("attack_type" , axis=1)
y=df["attack_type"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print(classification_report(y_pred, y_test))
print("confusion matrix: \n ", confusion_matrix(y_pred, y_test))
```

```python
penalty = ["l1", "l2", "elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga", "newton-cholesky"]
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator = model, param_grid = params, cv=cv, scoring="accuracy",n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print(classification_report(y_pred, y_test))
print("confusion matrix: \n ", confusion_matrix(y_pred, y_test))
```

```python
#one vs rest
```

```python
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
```

```python
onevsonemodel = OneVsOneClassifier(LogisticRegression()) 
```

```python
onevsrestmodel = OneVsRestClassifier(LogisticRegression())
```

```python
#onevsone : tüm ikili eşleşmeler için model oluşturur
onevsonemodel.fit(X_train, y_train) #daha iyi sonuç verdi
y_pred = onevsonemodel.predict(X_test)
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print(classification_report(y_pred, y_test))
print("confusion matrix: \n ", confusion_matrix(y_pred, y_test))
```

```python
#onevsrest : her sınıf için o sınıf ve diğer tüm sınıflara karşı model oluşturur
onevsrestmodel.fit(X_train, y_train)
y_pred = onevsrestmodel.predict(X_test)
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print(classification_report(y_pred, y_test))
print("confusion matrix: \n ", confusion_matrix(y_pred, y_test))
```
