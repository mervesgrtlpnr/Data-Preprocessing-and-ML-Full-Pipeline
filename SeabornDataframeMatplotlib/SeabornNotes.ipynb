```python
plt.scatter("Height" , "Weight" , data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") #se_style :arka plan stili, whitegrid:beyaz bir arka plan üzerine ızgara çizgileri
sns.scatterplot(x='Height',y='Weight', data=data) #dağılım grafiği (scatter plot)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex", data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
data["Medal"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Medal", data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x='Height',y='Weight', hue="Sex", data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x = "Height" , hue ="Sex" , data = data)
plt.ylabel("Frequency")
plt.title("Height Distribution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x = "Height", data = data)
plt.ylabel("Frequency")
plt.title("Height Distribution")
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x = "Height" , hue ="Sex" , data = data , kind = "kde") #dağılımı gösterdi
plt.ylabel("Frequency")
plt.title("Height Distribution")
plt.show()
```

```python
sns.barplot(x='Medal', y='Height', hue="Sex", data=data) #barlarla gösteriyor
plt.title('Medals by Height')
plt.show()
```

```python
sns.catplot(x='Medal', y='Height', hue="Sex",col="Season" , data=data) #kategorize edilmiş grafik
```
