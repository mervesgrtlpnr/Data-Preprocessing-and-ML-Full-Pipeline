```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
plt.subplot(1,2,1) # 1 satırda, 2 grafik (2 sütun) olacak , 1.sütun)
plt.plot(numpy_arr1,numpy_arr2,"r*-")
plt.subplot(1,2,2) #1 satırda, 2 grafik (2 sütun) olacak , 2.sütun)
plt.plot(numpy_arr2,numpy_arr1,"g--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3])#Eksenin sol kenarının, figürün sol kenarına olan uzaklığı,Eksenin alt kenarının, figürün alt kenarına olan uzaklığı,genişliği,yüksekliği
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
```

```python
new_fig = plt.figure(dpi=100)

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1,numpy_arr1 ** 2,label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3, label= "numpy array **3 ")
new_axes.legend(loc=6)#loc = location / legend çizgilerin neyi ifade ettiğini gösterir
```

```python
data1 = np.linspace(0,10,20)
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
type(my_fig)
```

```python
type(my_axes)
```

```python
my_fig , my_axes = plt.subplots()
my_axes.plot(data1, data2, color="#3A95A8",alpha=0.3) #alpha:transparanlık,doygunluk
my_axes.plot(data2, data1, color="#C96F23")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color="blue", linewidth = 1.0) #linewidth = kalınlık
new_axes.plot(data1, data1 + 3, color ="yellow", linewidth =4.0)

new_axes.plot(data1, data1 + 4, color = "#C96F23", linestyle = "-.")
new_axes.plot(data1, data1 + 5, color = "#C96F23", linestyle = ":")
new_axes.plot(data1, data1 + 6, color = "#C96F23", linestyle = "--")

new_axes.plot(data1, data1 + 7, color = "#000000", linestyle = "--", marker = "o",markersize = 8, markerfacecolor="red") # marker= veri noktalarını işaretleme / markersize = veri noktasının boyutu / rengi
new_axes.plot(data1, data1 + 8, color = "#000000", linestyle = "--", marker = "+",markersize = 4)
```
